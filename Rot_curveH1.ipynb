```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

import latex
import scienceplots
plt.style.use(['science','notebook', 'grid'])

import numpy as np
import pandas as pd


# Suppress warnings. Comment this out if you wish to see the warning messages
import warnings
warnings.filterwarnings('ignore')
```

```python
GalLong = np.array([15,25,35,45,55,65,75,85]) # galactic longitudes
#Dont include data for l > 90 deg longitudes
Rsun = 8.5 #kpc Distance of sun from galactic center
Vsun = 220 #km/s circular velcoity of sun about galactic center
```

$$V_{obs} = v_{source} - V_{sun} \sin{l} $$
$$ v_{source} = V_{obs} + V_{sun} \sin{l}$$

$v_{obs}$ is the (observed) rightmost (maximum) value of peak in corrected velocity v/s T_brightness plot

$R$ is the radial distance of the source from Galactic center. 
$$R = R_{sun}  \sin{l}$$

```python
v_obs = np.array([21.6290259807832, 39.2160962088244, 56.989793741571, 12.6001160031014, 5.14033233755057, 0.924089925910831, 6.01505619918679, 4.70907321255232])

# get 8 values of v_obs corresponding to 8 longitudes given, after multipeak fitting
```

```python
vSource = v_obs + Vsun*np.sin(np.deg2rad(GalLong))
R = Rsun*np.sin(np.deg2rad(GalLong))
```

```python
R
```

```python
plt.figure(figsize=(8, 5))
plt.title('HI Rotation Curve of Milky Way')
plt.xlim((0,10))
plt.ylim((50,250))
plt.xlabel('Distance from galactic center $R$ [kpc]',fontsize = 12)
plt.ylabel(r'Orbital Speed $v_{source}$ [km/s]',fontsize = 12)
plt.tick_params(axis='both', which='major', labelsize=10)  # For major ticks
plt.scatter(R,vSource,c='k')

plt.savefig('rot_curve.png',dpi=500,facecolor='w')
plt.show()
```
